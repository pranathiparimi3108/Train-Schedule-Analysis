```python
import pandas as pd
df = pd.read_csv('Dataset1.csv')
df["Arrival_time"] = pd.to_datetime(
    df["Arrival_time"],
    format="%H:%M:%S")
df["Departure_Time"] = pd.to_datetime(
    df["Departure_Time"],
    format="%H:%M:%S")
route_distance = df.groupby("Train_No")["Distance"].max()
print("Top 10 Longest Routes")
print(route_distance.sort_values(ascending=False).head(10))
print("\nRoute Categories")
for train, distance in route_distance.head(20).items():
  if distance < 300:
        category = "Short"
  elif distance < 700:
        category = "Medium"
  else:
        category = "Long"
  print(train, "-", category)
print("\nTop Stations")
print(df["Station_Name"].value_counts().head(10))
```
