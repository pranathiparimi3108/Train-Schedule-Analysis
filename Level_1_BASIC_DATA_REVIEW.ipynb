```python
import pandas as pd

df = pd.read_csv('Dataset1.csv')

print("Dataset Shape")
print(df.shape)

print("\nColumns")
print(df.columns)

print("\nFirst Five Records")
print(df.head())

print("\nTotal Trains")
print(df["Train_No"].nunique())

print("\nTotal Stations")
print(df["Station_Name"].nunique())

print("\nSample Routes")

for train in df["Train_No"].unique()[:10]:
    temp = df[df["Train_No"] == train]

    start_station = temp.iloc[0]["Station_Name"]
    end_station = temp.iloc[-1]["Station_Name"]

    print(train, ":", start_station, "->", end_station)

stops = df.groupby("Train_No")["Station_Name"].count()

print("\nMaximum Stops Train")
print(stops.idxmax(), "-", stops.max())

print("\nMinimum Stops Train")
print(stops.idxmin(), "-", stops.min())
```
