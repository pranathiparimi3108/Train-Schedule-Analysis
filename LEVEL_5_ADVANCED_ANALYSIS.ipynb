```python
import pandas as pd
df = pd.read_csv('Dataset1.csv')
pivot = pd.pivot_table(
    df,
    values="Train_No",
    index="Station_Name",
    aggfunc="count")
print("Station Wise Distribution")
print(pivot.sort_values(
        "Train_No",
        ascending=False
    ).head(20))
cross = pd.crosstab(
    df["Route_Number"],
    df["Station_Code"])
print("\nCross Tabulation")
print(cross.head())
```
