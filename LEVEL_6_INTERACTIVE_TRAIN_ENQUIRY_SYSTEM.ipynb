```python
import pandas as pd
df = pd.read_csv('Dataset1.csv')
source = input("Enter Source Station Code: ").upper()
destination = input("Enter Destination Station Code: ").upper()
found = False
for train in df["Train_No"].unique():
    temp = df[df["Train_No"] == train]
    stations = list(temp["Station_Code"])
    if source in stations and destination in stations:
        s = stations.index(source)
        d = stations.index(destination)
        if s < d:
            distance = (
                temp.iloc[d]["Distance"]
                - temp.iloc[s]["Distance"])
            print("\nTrain Found")
            print("Train Number:", train)
            print(
                "Source:",
                temp.iloc[s]["Station_Name"])
            print(
                "Destination:",
                temp.iloc[d]["Station_Name"])
            print("Distance:", distance, "KM")
            found = True
if found == False:
    print("\nNo Direct Train Found")
```
