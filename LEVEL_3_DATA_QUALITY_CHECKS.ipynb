```python
import pandas as pd
df = pd.read_csv('Dataset1.csv')
print("Missing Values")
print(df.isnull().sum())
duplicates = df.duplicated().sum()
print("\nDuplicate Rows")
print(duplicates)
df = df.drop_duplicates()
print("\nDataset Shape After Cleaning")
print(df.shape)
print("\nDistance Validation")
for train in df["Train_No"].unique()[:20]:
    temp = df[df["Train_No"] == train]
    if temp["Distance"].is_monotonic_increasing:
        print(train, "- Correct")
    else:
        print(train, "- Check")
df.to_csv("cleaned_train_data.csv", index=False)
print("\nCleaned File Saved")
```
