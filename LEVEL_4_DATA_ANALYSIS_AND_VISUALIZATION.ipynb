```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Dataset1.csv')
top_stations = df["Station_Name"].value_counts().head(10)
plt.figure(figsize=(10,5))
top_stations.plot(kind="bar")
plt.title("Top 10 Busy Stations")
plt.xlabel("Station")
plt.ylabel("Train Count")
plt.show()
route_distance = df.groupby("Train_No")["Distance"].max()
plt.figure(figsize=(10,5))
route_distance.head(20).plot(kind="bar")
plt.title("Route Distance")
plt.xlabel("Train Number")
plt.ylabel("Distance")
plt.show()
```
